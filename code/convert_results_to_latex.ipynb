```python
with open("results/cifar100_deep.csv", "r") as f:
    u = f.readlines()
```

```python
kl = u[2].split('\t')
kl2 = u[3].split('\t')
klk = u[4].split('\t')
kl2k = u[5].split('\t')
bcrm = u[6].split('\t')
logpl = u[7].split('\t')
```

```python
lines = template.split("\\\\")
all_lines = []
s = 0
for line in lines:
    parsed_line = line.strip()
    if parsed_line.startswith("\\mid"):
        print("MISC LINE: ", parsed_line)
        all_lines.append(parsed_line)
    else:
        parts = [item.strip() for item in parsed_line.split("&")]
        idx = (4 - (s % 5)) + 5 * (s // 5)
        # print(parts)
        parts[3] = '$' + kl[idx].replace("/", "\\pm") + '$'
        parts[4] = '$' + kl2[idx].replace("/", "\\pm") + '$'
        parts[5] = '$' + klk[idx].replace("/", "\\pm") + '$'
        parts[6] = '$' + kl2k[idx].replace("/", "\\pm") + '$'
        parts[7] = '$' + bcrm[idx].replace("/", "\\pm") + '$'
        if s % 5 == 0:
            parts[8] = "\\multirow{5}{*}{$" + str(logpl[idx]) + "$}" 
        all_lines.append(" & ".join(parts))
        s += 1
all_lines = " \\\\\n".join(all_lines)
```
