```python
import matplotlib.pyplot as plt
import statistics
from math import sqrt

```

```python
means = [np.mean([10, 11, 42, 45, 44]), np.mean([10, 21, 42, 45, 44]), np.mean([20, 2, 4, 45, 44]), np.mean([30, 31, 42, 45, 44])]
stds = [np.std([10, 11, 42, 45, 44]), np.std([10, 21, 42, 45, 44]), np.std([20, 2, 4, 45, 44]), np.std([30, 31, 42, 45, 44])]
n = len([10, 11, 42, 45, 44])
```

```python

def plot_w_int(means, stds, n, legend):
    def plot_confidence_interval(x, mean, stdev, n, z=1.96, color='gray', horizontal_line_width=0.25):
        confidence_interval = z * stdev / sqrt(n)

        left = x - horizontal_line_width / 2
        top = mean - confidence_interval
        right = x + horizontal_line_width / 2
        bottom = mean + confidence_interval
        plt.plot([x, x], [top, bottom], color=color)
        plt.plot([left, right], [top, top], color=color)
        plt.plot([left, right], [bottom, bottom], color=color)
        plt.plot(x, mean, 'o', color='red')

        return mean, confidence_interval
    plot_confidence_interval(1, means[0], stds[0], n) 
    plot_confidence_interval(5, means[1], stds[1], n)
    plot_confidence_interval(10, means[2], stds[2], n)
    plot_confidence_interval(25, means[3], stds[3], n)
    plot_confidence_interval(50, means[4], stds[4], n)
    plt.plot([1, 5, 10, 25, 50], means, label=legend)
```

```python
n = 3
```

```python
plt.figure(figsize=(8, 5))
plt.xticks([1, 5, 10, 25, 50], [1000, 5000, 10000, 25000, 50000])
plt.xlim([0, 55])
plt.xlabel('Number of logged known-reward samples', fontsize=16)
plt.ylabel('Accuracy (%)', fontsize=16)
plt.title('CIFAR-10', fontsize=20)
means = list(reversed([70.11666667, 58.33666667, 30.02666667, 24.68, 15.81333333]))
stds = list(reversed([6.888253286, 8.685720081, 12.75241676, 3.739465202, 5.11763943]))
plot_w_int(means, stds, n, legend='BanditNet')
means = list(reversed([69.53, 64.24, 63.19, 58.31333333, 51.136666673]))
stds = list(reversed([1.087244223, 2.090645833, 0.5109794516, 0.5208006656, 2.250185178]))
plot_w_int(means, stds, n, legend='KL-S2BL')
means = list(reversed([77.52666667, 76.46, 76.95666667, 74.13, 73.57333333]))
stds = list(reversed([1.189719855, 0.7014983963, 0.3527510926, 1.430908802, 0.3572580767]))
plot_w_int(means, stds, n, legend='WCE-S2BL')
plt.legend()
plt.grid()
# TODO: Export eps
```
